```python
def cyk_parser(grammar, string):
    n = len(string)

    # Create empty table (n x n)
    table = [[set() for _ in range(n)] for _ in range(n)]

    # Step 1: Fill diagonal
    for i in range(n):
        for lhs in grammar:
            for rhs in grammar[lhs]:
                if rhs == string[i]:
                    table[i][i].add(lhs)

    # Step 2: Fill upper triangle
    for length in range(2, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            for k in range(i, j):
                for lhs in grammar:
                    for rhs in grammar[lhs]:
                        if len(rhs) == 2:
                            B, C = rhs[0], rhs[1]
                            if B in table[i][k] and C in table[k + 1][j]:
                                table[i][j].add(lhs)

    return table


grammar = {
    'S': ['AB', 'BC'],
    'A': ['BA', 'a'],
    'B': ['CC', 'b'],
    'C': ['AB', 'a']
}

string = "baaba"

table = cyk_parser(grammar, string)


n = len(string)
print("\nCYK Parsing Table:\n")

for i in range(n):
    for j in range(n):
        if j < i:
            print("   ", end=" ")  # empty lower triangle
        else:
            print(f"{table[i][j]}", end=" ")
    print()

if 'S' in table[0][n - 1]:
    print("\nString is ACCEPTED by the grammar.")
else:
    print("\nString is NOT accepted by the grammar.")
```
