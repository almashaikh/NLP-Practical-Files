```python
def minDistance(word1: str, word2: str):
    n = len(word1)
    m = len(word2)

    # dp[i][j] = (cost, insertions, deletions, substitutions)
    dp = [[(0, 0, 0, 0) for _ in range(m + 1)] for _ in range(n + 1)]

    # Base cases
    for j in range(1, m + 1):
        # All insertions
        dp[0][j] = (j * 1, j, 0, 0)

    for i in range(1, n + 1):
        # All deletions
        dp[i][0] = (i * 1, 0, i, 0)

    # Fill DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):

            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                # Insert
                ins_cost, ins_i, ins_d, ins_s = dp[i][j - 1]
                insert = (ins_cost + 1, ins_i + 1, ins_d, ins_s)

                # Delete
                del_cost, del_i, del_d, del_s = dp[i - 1][j]
                delete = (del_cost + 1, del_i, del_d + 1, del_s)

                # Substitute
                sub_cost, sub_i, sub_d, sub_s = dp[i - 1][j - 1]
                substitute = (sub_cost + 2, sub_i, sub_d, sub_s + 1)

                # Choose minimum cost operation
                dp[i][j] = min(insert, delete, substitute, key=lambda x: x[0])

    return dp[n][m]


def main():
    word1 = input("Enter first word: ")
    word2 = input("Enter second word: ")

    cost, insertions, deletions, substitutions = minDistance(word1, word2)

    print("\nResults:")
    print("Minimum Cost:", cost)
    print("Insertions:", insertions)
    print("Deletions:", deletions)
    print("Substitutions:", substitutions)


if __name__ == "__main__":
    main()
```
