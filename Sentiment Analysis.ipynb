Single word Sentiment analysis

```python
from textblob import TextBlob

word = "good"
analysis = TextBlob(word)

print("Word:", word)
print("Polarity:", analysis.sentiment.polarity)
```

Multi-word Sentiment Analysis

```python
from textblob import TextBlob

sentence = "The movie was really amazing and enjoyable"
analysis = TextBlob(sentence)

print("Sentence:", sentence)
print("Polarity:", analysis.sentiment.polarity)
print("Subjectivity:", analysis.sentiment.subjectivity)
```

Polarity-Based Classification

```python
from textblob import TextBlob

text = "The product is not good at all"
analysis = TextBlob(text)

polarity = analysis.sentiment.polarity

if polarity > 0:
    sentiment = "Positive"
elif polarity < 0:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

print("Text:", text)
print("Polarity:", polarity)
print("Sentiment:", sentiment)
```

```python
texts = [
    "good",
    "bad",
    "The service was excellent",
    "I hate this product",
    "It is okay"
]

from textblob import TextBlob

for t in texts:
    analysis = TextBlob(t)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"

    print(f"{t} → {sentiment} (Polarity: {polarity})")
```
